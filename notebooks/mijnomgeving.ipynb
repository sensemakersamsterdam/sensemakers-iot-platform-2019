## Remove outliers

```python
# Remove obvious outliers by looking at the distribution.

# Plot a histogram with 50 bins, with a log-y scale.
plt.hist(df2["waterTemperature"], 50)
plt.yscale('log')
```

```python
# Remove obvious outliers by looking at the distribution.

# Plot a histogram with 50 bins, with a log-y scale.
plt.hist(df2["waterEC"], 50)
plt.yscale('log')
```

```python
# Remove data that is obviously wrong:
# - Temperature > 20 is probably from the sensors directly exposed to sunlight or inside a room.
# - Temperature = 0 probably means that the sensor did not measure properly.

df3 = df2
df3 = df3[df3["waterTemperature"] < 20]
df3 = df3[df3["waterTemperature"] > 0]
df3 = df3[df3["waterEC"] > 100]
df3 = df3[df3["waterEC"] < 8000]
print(f"Removed {len(df2)-len(df3)} outliers out of {len(df2)} points.")
```

```python
# Let us do some further cleaning of the data.
#
# We will use the aggregate values to find outliers.
# The idea is that we will only keep those points where the value is similar to the calculated mean over past 2 hours.

# For each point in time, calculate some aggregate values over the previous 2 hours.
# We will count the number of points, calculate the mean and the standard deviation.
count_2h = df3.groupby('name').rolling('2h').count().reset_index(0)
mean_2h = df3.groupby('name').rolling('2h').mean().reset_index(0)
std_2h = df3.groupby('name').rolling('2h').std().reset_index(0)
```

```python
# Add the aggregate values into the dataframe.
df3 = df3.join(count_2h, rsuffix='_count_2h')
df3 = df3.join(mean_2h, rsuffix='_mean_2h')
df3 = df3.join(std_2h, rsuffix='_std_2h')
```

```python
# Show the types of the columns.
df3.dtypes
```

```python
# We will use the aggregate values to find outliers.
# The idea is that we will only keep those points where the value is similar to the calculated mean over past 2 hours.
# 
# First, let us check if there was enough values to calculate the aggregate values.
# In principle, the measurements should come every 15 minutes, i.e.
# we should have 8 measurements available to calculate the 2-hours' mean.
# We will check this by looking at the histogram of counts.

# Plot a histogram with 16 bins from 0 to 15, with a log-y scale.
plt.hist(df3["waterTemperature_count_2h"], 16, (-0.5, 15.5))
plt.yscale('log')
```

```python
# Get rid of the points with less than 8 measurements per 2 hours.
print(f"Starting with {len(df3)} points.")
df3 = df3[df3["waterTemperature_count_2h"]>=8]
print(f"{len(df3)} points are kept after removing the points with less than 8 measurements per 2 hours.")
```

```python
print(f"Number of points with sigma = 0: {len(df3[df3['waterTemperature_std_2h']==0])}")
print(f"Number of points with sigma = 0: {len(df3[df3['waterEC_std_2h']==0])}")
df3 = df3[(df3["waterTemperature_std_2h"]>0) & (df3["waterEC_std_2h"]>0)]
print(f"Keeping {len(df3)} points.")
```

```python
# Let us look at the normalized differences from the 2-hours' mean.

x = (df3["waterTemperature"] - df3["waterTemperature_mean_2h"]) / df3["waterTemperature_std_2h"]

# Plot a histogram with 50 bins from -3 to 3.
plt.hist(x, 50, (-3, 3));

print(f"Number of points more than 2 sigma from the mean: {len(x[abs(x)>2])}")
df3 = df3[abs(x) < 2]
```

```python
# Let us look at the normalized differences from the 2-hours' mean.

x = (df3["waterEC"] - df3["waterEC_mean_2h"]) / df3["waterEC_std_2h"]

# Plot a histogram with 50 bins from -3 to 3.
plt.hist(x, 50, (-3, 3));

print(f"Number of points more than 2 sigma from the mean: {len(x[abs(x)>2])}")
df3 = df3[abs(x) < 2]
```

## Analysis

```python
# In the following, we will work with hourly means and look at the differences.

# Prepare a dataset with hourly means.
df4 = df3.groupby('name')[['waterTemperature', 'waterEC', 'name']].resample('H').mean().reset_index(0)
df4.head()
```

```python
# Calculate differences between two consecutive points.
# In other words, extract the changes (in temperature and EC) over one hour.
df4_diff = df4.groupby('name').diff()
df4_diff['name'] = df4['name']

print("Temperature standard deviation:", df4_diff["waterTemperature"].std())
print("EC standard deviation:", df4_diff["waterEC"].std())

# We will set alerts based on 3-sigma deviations (rounded off to nice values).
T3S = 0.5
E3S = 250
T2S = T3S/3*2
E2S = E3S/3*2
```

```python
# Overflow and underflow bins
df4_diff["waterTemperature"] = df4_diff["waterTemperature"].clip(-3*T3S, 3*T3S)
df4_diff["waterEC"] = df4_diff["waterEC"].clip(-3*E3S, 3*E3S)

# Show the changes over one hour in histograms.
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(10,5))
ax0.hist(df4_diff['waterTemperature'], 50, (-3*T3S, 3*T3S))
ax0.set_xlim(-3*T3S, 3*T3S)
ax0.set_yscale('log')
ax0.set_xlabel("Temperature")
ax0.axvline(x=-T3S, color='r')
ax0.axvline(x=T3S, color='r', label='$\pm3\sigma$')
ax0.legend()

ax1.hist(df4_diff['waterEC'], 50, (-3*E3S, 3*E3S))
ax1.set_xlim(-3*E3S, 3*E3S)
ax1.set_yscale('log')
ax1.set_xlabel("EC");
ax1.axvline(x=-E3S, color='r')
ax1.axvline(x=E3S, color='r', label='$\pm3\sigma$')
ax1.legend();
```

```python
import matplotlib.colors as colors

x = df4_diff["waterTemperature"].to_list()
y = df4_diff["waterEC"].to_list()

fig, ax = plt.subplots()
h = ax.hist2d(x, y, bins=25, range=[[-3*T3S, 3*T3S], [-3*E3S, 3*E3S]], norm=colors.LogNorm())
plt.colorbar(h[3], ax=ax);
ax.set_xlim(-3*T3S, 3*T3S)
ax.set_ylim(-3*E3S, 3*E3S)
ax.set_xlabel("Temperature$-\mu$")
ax.set_ylabel("EC$-\mu$")
ax.axvline(x=-T2S, ymin=0, ymax=0.39, color='r')
ax.axvline(x=-T2S, ymin=0.61, ymax=1, color='r')
ax.axvline(x=T2S, ymin=0, ymax=0.39, color='r')
ax.axvline(x=T2S, ymin=0.61, ymax=1, color='r')
ax.axhline(y=-E2S, xmin=0, xmax=0.39, color='r')
ax.axhline(y=-E2S, xmin=0.61, xmax=1, color='r')
ax.axhline(y=E2S, xmin=0, xmax=0.39, color='r')
ax.axhline(y=E2S, xmin=0.61, xmax=1, color='r', label='$\pm2\sigma$')
ax.legend();
```

```python
# The threshold values for alerting are chosen roughly at 3 sigma.
alert_temp = abs(df4_diff["waterTemperature"]) > T3S
alert_EC = abs(df4_diff["waterEC"]) > E3S
alert_both = alert_temp & alert_EC
#alert_both = (abs(df4_diff["waterTemperature"]) > (T3S/3.*2.)) & (abs(df4_diff["waterEC"]) > (E3S/3.*2.))
```

```python
alert = df4[['name']]
alert['alert_temp'] = alert_temp
alert['alert_EC'] = alert_EC
alert['alert_both'] = alert_both
alert
```

```python
# Show a heatmap of hourly temperatures for all sensors in time.
data = df4.groupby(['name', 'datetime']).max()['waterTemperature'].unstack()
x_index = data.columns
y_index = data.index.values

data = data.to_numpy(dtype='float')
data = np.clip(data, 0, 20)

fig = plt.figure(figsize=(15,15))
plt.title("Temperature")
ax = fig.add_subplot(111)
im = ax.imshow(data, cmap='jet', interpolation='nearest', aspect='auto', vmin=0, vmax=20)

ax.set_xticks(np.arange(len(x_index), step=24))
ax.set_xticklabels(x_index[::24], rotation=30)
ax.set_yticks(np.arange(len(y_index)))
ax.set_yticklabels(y_index)
plt.colorbar(im);
```
